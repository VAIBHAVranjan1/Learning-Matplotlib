# FIFA

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fifa = pd.read_csv('./fifa_data.csv')
```

```python
fifa.head(10)
```

## Histogram

```python
bins = np.arange(40, 110, 10)
plt.hist(fifa.Overall, bins=bins, color='#a6e693')
plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')
plt.show()
```

## PIE CHART

```python
left = len(fifa.loc[fifa['Preferred Foot'] == 'Left'])
right = len(fifa.loc[fifa['Preferred Foot'] == 'Right'])

plt.figure(figsize=(5, 4), dpi=800)
plt.pie([left, right], labels=["left foot", "right foot"], hatch=['.' , '|'], autopct='%.3f %%', labeldistance=1.5)
plt.show()
```

## 2nd Pie Chart

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

light_weight = len(fifa.loc[fifa['Weight'] <= 125])
light_medium = len(fifa.loc[(fifa['Weight'] > 125) & (fifa['Weight'] <= 150)])
medium = len(fifa.loc[(fifa['Weight'] > 150) & (fifa['Weight'] <= 175)])
medium_heavy = len(fifa.loc[(fifa['Weight'] > 175) & (fifa['Weight'] <= 200)])
heavy_weight = len(fifa.loc[fifa['Weight'] > 200])

weights = [light_weight, light_medium, medium, medium_heavy, heavy_weight]
labels = ["Under 125", "125 - 150", "150 - 170", "170 - 200", "Over 200"]
```

```python
plt.style.use('ggplot')
explode = (0.5,0.1,0.1,0.1,0.5, )

plt.title("Weight Distribution Of Fifa (in lbs)")
plt.pie(weights, labels=labels, autopct='%.2f%%', pctdistance=0.8, explode=explode)
plt.show()
```

```python
# All teams

top_players = fifa.loc[fifa.Overall >= 90, ['Name', 'Overall']]
print(top_players)
```

```python
plt.style.use('default')
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
new_england = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

```

## Box Plot

```python
plt.figure(figsize=(5, 8))
labels = ['Barcelona', 'Real Madrid', 'New England Revolution']
box = plt.boxplot([barcelona, madrid, new_england], label=labels, patch_artist=True)

for boxes in box['boxes']:
    boxes.set(color = '#ed112e', linewidth=2, facecolor='#84db67')

plt.title("Proffessional Soccer Team Comparison")
plt.ylabel("Overall Rating")
plt.show()
```

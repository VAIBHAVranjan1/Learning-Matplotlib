### Functional Plot

```python
x_1 = np.linspace(0, 5, 10)
y_1 = x_1**2
plt.plot(x_1, y_1)
plt.title("Days Squared Charts")
plt.xlabel("Days")
plt.ylabel("Days Squared")
```

### Print Multiple Plots

```python
# Subplot() is used to create multiple plts in a single frame
# It allows you to organize your plots in a grid layout (like rows and columns)
# possilbe values are (1, 1), (1, 2), (2, 1), (2, 2)

# Plot 1 (top-left)
plt.subplot(3, 3, 1)
plt.plot(x_1, y_1, 'g')
plt.title("Left Plot")

# Plot 2 (top-middle)
plt.subplot(3, 3, 2)
plt.plot(x_1, y_1, 'b')
plt.title("Middle Plot")

# Plot 3 (top-right)
plt.subplot(3, 3, 3)
plt.plot(x_1, y_1, 'y')
plt.title("Right Plot")

# Plot 4 (middle-left)
plt.subplot(3, 3, 4)
plt.plot(x_1, y_1, 'r')
plt.title("middle Leftmost Plot")

# lowest leftmost 
plt.subplot(3, 3, 7)
plt.plot(x_1, y_1, 'k')
plt.title("Lowest Leftmost Plot")

# plt.subplot(rows, cols, index) → grid position from left to right, top to bottom.
# index goes from 1 -> (rows * cols)

plt.tight_layout() 
# this command will spread equally
```

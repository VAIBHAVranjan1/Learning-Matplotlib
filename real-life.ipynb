# Real Life Examples

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('./gas_prices.csv')
```

```python
countries = gas.columns[1:].tolist()

plt.figure(figsize = (16, 9))
for country in countries:
    plt.plot(gas.Year, gas[country], label=f'{country}', marker='.')
plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Gas Prices by Country Over Time', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.legend()
plt.savefig('gasvscountries.png', dpi = 1000)
```
